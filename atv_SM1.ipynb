```python
# Leia o salário de um funcionário, Até 2000 → aumento de 10%, Acima de 2000 → aumento de 5%

salario = float(input("Digite seu salário: R$"))
if salario <= 2000:
  novo_salario = salario + (salario * 0.1)
  print("Seu novo salário com um aumento de 10% aplicado é de R$", novo_salario)
elif salario > 2000:
  novo_salario = salario + (salario * 0.05)
  print("Seu novo salário com um aumento de 5% aplicado é de R$", novo_salario)
else:
  print("Salário inválido.")
```

```python
#Leia o valor de um produto, Se for maior que 100 → desconto de 10%, Caso contrário → desconto de 5%

valor = float(input("Digite o valor do produto: R$"))
if valor <= 100:
  novo_valor = valor - (valor * 0.05)
  print("Novo valor do produto com um desconto de 5% aplicado: R$", novo_valor)
elif valor > 100:
  novo_valor = valor - (valor * 0.1)
  print("Novo valor do produto com um desconto de 10% aplicado: R$", novo_valor)
else:
  print("Valor inválido.")
```

```python
#Leia a, b e c e informe se a equação tem: 2 raízes reais, 1 raiz real ou nenhuma raiz real

a = float(input("Digite o valor de a: "))
b = float(input("Digite o valor de b: "))
c = float(input("Digite o valor de c: "))
delta = (b * b) - (4 * a * c)
if delta == 0:
  print("A equação tem uma raiz real.")
elif delta > 0:
  print("A equação tem duas raizes reais.")
elif delta < 0:
  print("A equação não tem raiz real.")
else:
  print("Delta inválido.")
```

```python
#Calcule as raízes da equação do 2º grau (considere apenas Δ ≥ 0).
import math

a = float(input("Digite o valor de a: "))
b = float(input("Digite o valor de b: "))
c = float(input("Digite o valor de c: "))
delta = (b * b) - (4 * a * c)

if delta == 0:
  x1 = (-b + math.sqrt(delta)) / (2*a)
  print("Uma raiz real: ", x1)
else:
  x1 = (-b + math.sqrt(delta)) / (2*a)
  x2 = (-b - math.sqrt(delta)) / (2*a)
  print(f"Duas raízes reais: {x1} e {x2}")
```
